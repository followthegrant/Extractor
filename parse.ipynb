```python
import csv
import os
from bs4 import BeautifulSoup as bs
```

```python
dir = 'data/papers/nature-com/american-journal-of-gastroenterology'
for filename in os.listdir(dir):
    #print(filename)

    with open(dir +'/'+ filename,'r') as file:
        soup = bs(file, 'lxml')
        #print(soup.find('title').text.split('|')[0])
        authors = [s['content'] for s in soup.find_all('meta', {'name':'dc.creator'})]
        #print(authors)
        title = soup.find('meta', {'name':'dc.title'})['content']
        #print(title)
        doi = soup.find('meta', {'name':'DOI'})['content']
        article_type = soup.find('meta', {'name':'dc.type'})['content']
        date = soup.find('meta', {'name':'prism.publicationDate'})['content']
        abstract = soup.find('meta', {'name':'dc.description'})['content']
        subject = soup.find('meta', {'name':'dc.subject'})['content']
        print(subject)
```
